# Quant OS - Trading Dashboard
Real-time monitoring for paper & live trading

```python
import sys, os, asyncio, json
from datetime import datetime
from decimal import Decimal
from pathlib import Path

sys.path.insert(0, os.getcwd())

import MetaTrader5 as mt5
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

from graxia.packages.quant_os.core.config import get_config
from graxia.packages.quant_os.risk.kill_switch import KillSwitch
from graxia.packages.quant_os.risk.circuit_breaker import CircuitBreaker

mt5_ok = mt5.initialize()
account = mt5.account_info() if mt5_ok else None
print(f'MT5: {"Connected" if mt5_ok else "Failed"}')
if account:
    print(f'Account: {account.login} ({account.server})')
    print(f'Balance: \${account.balance:,.2f} | Equity: \${account.equity:,.2f}')
```

```python
def account_card():
    a = mt5.account_info()
    if not a: return HTML('<p>MT5 disconnected</p>')
    pnl = a.equity - a.balance
    pnl_color = 'green' if pnl >= 0 else 'red'
    return HTML(f'''
    <div style="border:1px solid #ccc; padding:15px; border-radius:8px; margin:10px 0;">
        <h3>{a.name}</h3>
        <table>
        <tr><td>Account</td><td><b>{a.login}</b></td><td>Server</td><td><b>{a.server}</b></td></tr>
        <tr><td>Balance</td><td><b>\${a.balance:,.2f}</b></td><td>Equity</td><td><b>\${a.equity:,.2f}</b></td></tr>
        <tr><td>P&L</td><td><b style="color:{pnl_color}">\${pnl:+,.2f}</b></td><td>Leverage</td><td><b>1:{a.leverage}</b></td></tr>
        <tr><td>Free Margin</td><td><b>\${a.margin_free:,.2f}</b></td><td>Currency</td><td><b>{a.currency}</b></td></tr>
        </table>
    </div>''')

display(account_card())
```

```python
def positions_table():
    positions = mt5.positions_get() if mt5_ok else None
    if not positions:
        return HTML('<p>No open positions</p>')
    rows = []
    for p in positions:
        pnl_color = 'green' if p.profit >= 0 else 'red'
        rows.append(f'''<tr>
            <td>{p.symbol}</td>
            <td>{"BUY" if p.type == 0 else "SELL"}</td>
            <td>{p.volume:.2f}</td>
            <td>{p.price_open:.5f}</td>
            <td>{p.sl:.5f if p.sl else "-"}</td>
            <td>{p.tp:.5f if p.tp else "-"}</td>
            <td>{p.price_current:.5f}</td>
            <td style="color:{pnl_color}">\${p.profit:+,.2f}</td>
        </tr>''')
    
    return HTML(f'''
    <h3>Open Positions ({len(positions)})</h3>
    <table border="1" style="border-collapse:collapse; width:100%;">
        <tr><th>Symbol</th><th>Type</th><th>Volume</th><th>Open</th><th>SL</th><th>TP</th><th>Current</th><th>P&L</th></tr>
        {"".join(rows)}
    </table>''')

display(positions_table())
```

```python
def risk_status():
    ks = KillSwitch()
    cb = CircuitBreaker()
    config = get_config()
    ks_status = 'Active' if ks.is_active() else 'Armed'
    ks_color = 'red' if ks.is_active() else 'green'
    cb_status = 'Blocked' if cb.is_blocked else 'Closed'
    cb_color = 'red' if cb.is_blocked else 'green'
    return HTML(f'''
    <div style="border:1px solid #ccc; padding:15px; border-radius:8px; margin:10px 0;">
        <h3>Risk Status</h3>
        <table>
        <tr><td>Mode</td><td><b>{config.trading_mode.value}</b></td></tr>
        <tr><td>Kill Switch</td><td><b style="color:{ks_color}">{ks_status}</b></td></tr>
        <tr><td>Circuit Breaker</td><td><b style="color:{cb_color}">{cb_status}</b></td></tr>
        <tr><td>Max Risk/Trade</td><td><b>{config.max_risk_per_trade_pct}%</b></td></tr>
        <tr><td>Max Daily Loss</td><td><b>{config.max_daily_loss_pct}%</b></td></tr>
        <tr><td>Max Drawdown</td><td><b>{config.max_drawdown_pct}%</b></td></tr>
        </table>
    </div>''')

display(risk_status())
```

```python
def recent_orders():
    if not mt5_ok:
        return HTML('<p>MT5 not connected</p>')
    orders = mt5.history_deals_get(datetime(2020,1,1), datetime.now()) or []
    recent = sorted(orders, key=lambda x: x.time, reverse=True)[:10]
    if not recent:
        return HTML('<p>No recent orders</p>')
    rows = []
    for d in recent:
        rows.append(f'''<tr>
            <td>{d.symbol}</td>
            <td>{"BUY" if d.type == 0 else "SELL"}</td>
            <td>{d.volume:.2f}</td>
            <td>{d.price:.5f}</td>
            <td>\${d.profit:+,.2f}</td>
            <td>{d.comment or "-"}</td>
        </tr>''')
    return HTML(f'''
    <h3>Recent Orders (last 10)</h3>
    <table border="1" style="border-collapse:collapse; width:100%;">
        <tr><th>Symbol</th><th>Type</th><th>Volume</th><th>Price</th><th>P&L</th><th>Comment</th></tr>
        {"".join(rows)}
    </table>''')

display(recent_orders())
```

```python
import time
refresh_interval = widgets.IntSlider(value=5, min=1, max=30, description='Refresh (s):')
refresh_btn = widgets.Button(description='Start Auto-Refresh')
stop_btn = widgets.Button(description='Stop')
output = widgets.Output()

display(widgets.HBox([refresh_interval, refresh_btn, stop_btn]))
display(output)

auto_refresh = False

def start_refresh(b):
    global auto_refresh
    auto_refresh = True
    while auto_refresh:
        with output:
            clear_output(wait=True)
            display(account_card())
            display(positions_table())
            display(risk_status())
        time.sleep(refresh_interval.value)

def stop_refresh(b):
    global auto_refresh
    auto_refresh = False

refresh_btn.on_click(start_refresh)
stop_btn.on_click(stop_refresh)
print('Click Start Auto-Refresh to begin live updates')
```

```python
mt5.shutdown()
print('MT5 disconnected')
```
